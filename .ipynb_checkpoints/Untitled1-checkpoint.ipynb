```python
import pandas as pd
census_df = pd.read_csv('census.csv')
columns_to_keep = ['STNAME',
               'POPESTIMATE2010',
               'POPESTIMATE2011',
               'POPESTIMATE2012',
               'POPESTIMATE2013',
               'POPESTIMATE2014',
               'POPESTIMATE2015']
df1 = census_df.set_index('CTYNAME')
df1 = df1[columns_to_keep]
df1['Minimum'] = df1.loc[:,['POPESTIMATE2010','POPESTIMATE2011','POPESTIMATE2012','POPESTIMATE2013','POPESTIMATE2014','POPESTIMATE2015']].min(axis = 1)
df1['Maximun'] = df1.loc[:,['POPESTIMATE2010','POPESTIMATE2011','POPESTIMATE2012','POPESTIMATE2013','POPESTIMATE2014','POPESTIMATE2015']].max(axis = 1)
df1['abs_value'] = df1['Maximun'] - df1['Minimum']
final = df1['abs_value'].idxmax()
final
```
