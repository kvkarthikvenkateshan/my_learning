````python
import pandas as pd
import numpy as np
df = pd.read_csv('census.csv')
df.head(5)
````

````python
df = df.set_index('STNAME')
def fun(item):
    if item[0]<'M':
        return 0
    if item[0]<'Q':
        return 1
    return 2

for group, frame in df.groupby(fun):
    print('There are ' + str(len(frame)) + ' records in group ' + str(group) + ' for processing.')
````

In case of the above example also we are passing a function named 'fun' as the argument for the group by method and the index of the datatframe becomes the argument for the defined function i.e. item has all the state names

## Scales

Generally in DataScience we make use of 4 scales 

Ratio Scale: Units are equally spaced and mathematical operations such as +/- are all valid

    Ex: Height and weight

Interval Scale: Units are equally spaced, but there is no true Zero.
    
    Ex: Temperature and angle measured in degrees
    here temperature 0 doen't mean the absence of temperature and similarly 0 degree don't mean no angle.

Ordinal Scale: This scale is ordered but the units are not equally spaced.
    
    Ex: Grades in a college, the difference between an A grade and B grade is 15 while the difference between B grade               and c is 10

Nominal Scale: In this scale the units are not order and dont have equal interval between them.
    
    Ex: Tick boxes in a survey form asking for Gender, here the intension is to capture the data the order of Male or                 Female is irrespective

````python
import pandas as pd
import numpy as np
import re

def get_energy():
    energy = pd.read_excel('Energy Indicators.xls', sheetname = 'Energy', header = 15, skip_footer = 38)
    energy.drop(['Unnamed: 0','Unnamed: 1'], axis = 1, inplace = True)
    energy.rename(columns = {'Unnamed: 2' : 'Country', 'Renewable Electricity Production' : '% Renewable'},inplace = True)
    energy.drop(0, axis = 0, inplace = True)
    energy.replace('...', np.nan, inplace = True)

    val = '[a-zA-Z,\s]+'
    l = []
    for i in energy['Country']:
        res = re.search(val,i)
        s = res.group(0)
        l.append(s)
    energy['Country'] = l

    con_energy = energy['Energy Supply']
    con_energy *=  1000000
    energy.fillna(value = np.nan , inplace = True)
    energy.replace({"Republic of Korea": "South Korea",
                 "United States of America": "United States",
                 "United Kingdom of Great Britain and Northern Ireland": "United Kingdom",
                 "China, Hong Kong Special Administrative Region": "Hong Kong",
                 "Iran ": "Iran"}, inplace = True)
    energy = energy.set_index('Country')
    return energy

def get_GDP():
    GDP = pd.read_csv('world_bank.csv',header = 4)
    GDP.rename(columns = {'Country Name' : 'Country'}, inplace = True)
    GDP.replace({"Korea, Rep.": "South Korea", "Iran, Islamic Rep.": "Iran","Hong Kong SAR, China": "Hong Kong"}, inplace = True)
    GDP.drop(GDP.columns[1:50],axis = 1,inplace = True)
    GDP = GDP.set_index('Country')
    return GDP

def get_ScimEn():
    ScimEn = pd.read_excel('scimagojr-3.xlsx')
    ScimEn = ScimEn[ScimEn['Rank'] < 16]
    ScimEn =ScimEn.set_index('Country')
    return ScimEn

def answer_one():
    energy = get_energy()
    GDP = get_GDP()
    ScimEn = get_ScimEn()
    pre = pd.merge(ScimEn, energy, how = 'inner', left_index=True, right_index=True )
    final = pd.merge(pre, GDP, how = 'inner', left_index=True, right_index=True)
    final.sort(columns = 'Rank', axis = 0, ascending=True, inplace = True)
    return final

final = answer_one()
````

````python
chance_of_tornado = 0.01

tornado_events = np.random.binomial(1, chance_of_tornado, 1000000)
#print(len(tornado_events))
two_days_in_a_row = 0
for j in range(1,len(tornado_events)-1):
    if tornado_events[j]==1 and tornado_events[j-1]==1:
        two_days_in_a_row+=1

print('{} tornadoes back to back in {} years'.format(two_days_in_a_row, 1000000/365))
````
